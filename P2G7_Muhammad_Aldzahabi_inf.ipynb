```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from keras.applications.mobilenet_v2 import preprocess_input
```

```python
loaded_model = load_model("Skin_Project.keras")
```

# 1. Normal

```python
img_path = "Normal.jpg"

# Load & resize image
img = image.load_img(img_path, target_size=(224, 224))

plt.imshow(img)
plt.axis('off')
plt.title("Input Image")
plt.show()

# Convert ke array
img_array = image.img_to_array(img)

# Tambah batch dimension
img_array = np.expand_dims(img_array, axis=0)

# Preprocess sesuai MobileNetV2
img_array = preprocess_input(img_array)
```

```python
# Predict
pred_probs = loaded_model.predict(img_array)

# Ambil probabilitas (sigmoid output)
pred_prob = pred_probs[0][0]

# Threshold  class
pred_class = int(pred_prob >= 0.5)

# Class names 
class_names = ['Jerawat', 'Darkspot', 'Kemerahan', 'Kerutan', 'Normal']

print("Prediction:", class_names[pred_class])
print("Confidence:", pred_prob)
```
